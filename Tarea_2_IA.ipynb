```python
#Se genera un dataset con el mismo formato que el anterior, pero que abarque una semana en especifico
days = pd.date_range(start='2023-01-01', end='2023-12-31')
data_in_list = []
#Parámetro de semana
week = 1

for day in days:
  if day.isocalendar().week != week:
    continue
  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': 0, 'Formula 1': rd.randint(1, 3), 'Tenis': rd.randint(1, 3)}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': 0, 'Formula 1': 0, 'Tenis': rd.randint(1, 3)}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': 0, 'Formula 1': 0, 'Tenis': 0}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': 0, 'Formula 1': rd.randint(1, 3), 'Tenis': 0}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': rd.randint(1, 3), 'Formula 1': 0, 'Tenis': 0}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': rd.randint(1, 3), 'Formula 1': rd.randint(1, 3), 'Tenis': 0}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': rd.randint(1, 3), 'Formula 1': 0, 'Tenis': rd.randint(1, 3)}
  data_in_list.append(nueva_fila)

  nueva_fila = { 'Date': pd.to_datetime(day), 'Padel': rd.randint(1, 3), 'Formula 1': rd.randint(1, 3), 'Tenis': rd.randint(1, 3)}
  data_in_list.append(nueva_fila)

df = pd.DataFrame.from_records(data_in_list)

print(df)
```

```python
import datetime as dt
from datetime import datetime

#Con esta función se generan las partículas para el filtro, se reciben como parámetros el dataframe inicial (el grande)
#y el día para generar las partículas
def generar_particulas(df, dia):
    particulas = []
    years = [2017, 2018, 2019, 2020, 2021]
    for year in years:
        fecha_str = str(year)
        if dia.month < 10:
          fecha_str += "-0"+str(dia.month)
        else:
          fecha_str += "-"+str(dia.month)

        if dia.day < 10:
          fecha_str += "-0"+str(dia.day)
        else:
          fecha_str += "-"+str(dia.day)

        #Se filtra la información del dataset inicial por medio de la fecha.
        data = df.loc[df['Date'] == fecha_str, ['Padel', 'Formula 1', 'Tenis']]

        #El formato de la lista añadida es
        #[Apuestas realizadas en el día, porcentaje de similitud con la información observada en el filtro, fecha de la apuesta]
        particulas.append([data, 0, df.loc[df['Date'] == fecha_str, ['Date']].head(1)])
    return particulas

#Definir la función para aplicar el filtro de partículas
#df es el dataset de observación, week es la semana que se observará, iteraciones es la cantidad de veces
#que se realizará el filtrado y resampling_force es un parámetro para generar ruido
def apply_filter_of_particles(df, week, iteraciones, resampling_force):
    #Obtener los datos de entrenamiento
    train_data = pd.read_csv('dataset_apuestas.csv', parse_dates=['Date'])

    #Se filtran solo aquellos datos que pertenezcan a la semana en observación pero de los años anteriores
    train_data = train_data[train_data['Date'].dt.isocalendar().week == week]


    #Obtener las observaciones de la semana seleccionada en 2023
    test_data = df

    #la variable paso es para evitar el recalculo de distintas operaciones
    paso = 0

    #Aquí se guardará la información a retornar
    retorno = []

    for index, row in test_data.iterrows():
      #Si paso es igual a 0, se realiza el calculo, y si es 8, se reinicia la variable paso a 0
      if paso == 0:
        block = test_data.loc[test_data['Date'] == row['Date'], ['Padel', 'Formula 1', 'Tenis']]
        distribucion = generar_particulas(train_data, row['Date'])
        for i in range(iteraciones):
          new_dist = [] #esta distribución reemplazará a la actual al final de cada iteración (actualización bayesiana)
          total_w = 0 #peso total de las particulas
          actual_distribucion = distribucion #el nombre es descriptivo
          for dist in actual_distribucion: #por cada particula se calcula su peso y se normaliza
            comun = pd.merge(block, dist[0], how='inner')
            W = len(comun) #se cuentan los elementos en común (multiplicadores) entre la particula y la observación
            dist[1] = W
            total_w += W
          for dist in actual_distribucion: #resampling
            if total_w > 0:
              W_normalizado = dist[1]/total_w
            else:
              W_normalizado = 0
            dist[1] = W_normalizado
            if dist[1] >= resampling_force: #Si el peso normalizado es mayor o igual a resampling_force, se guarda en la nueva distribución
              new_dist.append(dist)
          actual_distribucion = new_dist #Actualización bayesiana
        retorno.append(actual_distribucion) #Se añade el resultado del filtrado para el día x de la semana.

      if paso == 8:
        paso = 0
        continue
      paso += 1

    return retorno

resultados = apply_filter_of_particles(df, week, 1, 0.4)

print("Para la siguiente semana:")
for day in days:
  if day.isocalendar().week == week:
    print(day)

print("\nSe puede respaldar la apuesta en base a los siguientes porcentajes de semejanza")

for resultado in resultados:
  for element in resultado:
    print("_-_-_-_-_-_-_-_-")
    print(str(element[1]*100)+"%\n"+str(element[2]['Date'].iloc[0])) #Se imprime la probabilidad de semejanza junto a la fecha correspondiente

```
